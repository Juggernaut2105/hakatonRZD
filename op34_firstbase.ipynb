```python
#возможно нужные стольбцы
colsimportant=['stan_op', 'kop_vmd','nom_vag','date_op','req_date_time','ont_time_write', 'dvs',  'kop_okpv',   
         'index_poezd', 'kds', 
        'kod_sost_uch',      'kop_pmd', 'kop_prib202',      'nom_poezd', 'nom_put',
        'tip_edv','vid_dis_stan', 'pr_op', 'pr_prom',  'pr_vag_poezd',
        ]

#наверно ненужные столбцы:
# colstodrop=['date_izm_nom','dornod_posl','dornod_prib202','dor_dis','dor_rasch', 'id','kod_nod','kod_sob','kpp_t','napr_posl',
#        'napr_prib202','nbe','nod_dis', 'nom_park','oc_dis', 'ppv_gruj', 'ppv_gruz_os', 'ppv_kkb', 'ppv_kon','ppv_mest', 'ppv_mest_nod','ppv_mo',
#        'ppv_nrp','ppv_por','ppv_rd','ppv_rp','ppv_svrf', 'ppv_tranz','ppv_tr_nod', 'ppv_xar_por','prob_grj', 'prob_por', 'prob_stan',
#        'pr_gtd', 'pr_kach','pr_styk_posl','pr_styk_prib202','rgn_dis',  'sluj_tip_edv', 'stan_dis', 'stan_nazn','uns','uns_t','vid_dis_dor',
#        'vnrp_iskl', 'vnrp_neisp',
#        'vnrp_ostal', 'vnrp_peres', 'vnrp_pod_iskl', 'vnrp_rezmps',
#        'vnrp_spec_tex', 'vnrp_zapmps' 'dvs_t','kds_t','kod_vida_sled','request_unq_seq','rod_vag_uch','nom_vag_op','nom_vag_st','nom_vagon_st', ]
df_=df[colsimportant]
```

```python
#сгруппировать по станции, номеру вагона и коду операции для сохранения начальных и конечных моментов времени
df_g=df_.groupby(['stan_op','kop_vmd','nom_vag']).agg(
date_op= pd.NamedAgg(column = 'date_op', aggfunc = 'min'),
req_date_time= pd.NamedAgg(column = 'req_date_time', aggfunc = 'min'),
ont_time_write = pd.NamedAgg(column = 'ont_time_write', aggfunc = 'max'))
df_g=df_g.reset_index()

#просто тест
# display(df_g[(df_g.stan_op==611000) & (df_g.nom_vag=='0001d6ad2ea5ffb29216abcf291fbcf1')],df_[(df_.stan_op==611000) & (df_.nom_vag=='0001d6ad2ea5ffb29216abcf291fbcf1')])
```

```python
#в базе много дублированных значений, цель мержа оставить строки с первым упоминанием операции 3 и последним упоминанием операции 4
dftmp=pd.merge(df_,df_g,how='inner',on=['stan_op','kop_vmd','nom_vag']) 
#далее оставить из всех операций только 3 и 4, причем с условием первого упоминания 3 и последнего упоминания 4
dftmp=dftmp[(dftmp.kop_vmd==3) & (dftmp.req_date_time_x==dftmp.req_date_time_y) & (dftmp.date_op_x==dftmp.date_op_y) | (dftmp.kop_vmd==4) & (dftmp.ont_time_write_x==dftmp.ont_time_write_y)]
#удалить с префиксом _y , оставшиеся от мержа - они нужны были только для фильтрации, отсортировать
#для отдельно взятого вагона операция 3 всегда идет раньше чем 4
dftmp=dftmp.drop(['date_op_y',	'req_date_time_y',	'ont_time_write_y'],axis=1).sort_values(['nom_vag','date_op_x'],ascending=True) #где в выборку попали события висключения из поезда и включения в поезд
dftmp[:5]
```

```python
#для для каждой строки из всей выборки взять следующую строку - в ней будет инфа по тому же вагону для операции 4, т.к. всего оставлено операции 3 и 4
dftmp[['stan_op_','kop_vmd_','date_op_x_','nom_vag_']]=dftmp.shift(-1)[['stan_op','kop_vmd','date_op_x','nom_vag']] #взять предыдущую строку
#выкинуть строки, слепленные из разных вагонов и/или станций, останутся только одинаковые станция/вагон
dftmp=dftmp.query('nom_vag==nom_vag_ & stan_op==stan_op_' ).drop(['nom_vag_','stan_op_'], axis=1) #удалить строки, где попались вагоны с операцией без пары (3 без 4 или наоборот) и убрать столбец с проверочным номером

#вот итоговый искомый результат - время от начала операции 3(kop_vmd) до конца операции 4(kop_vmd_)
#по каждой строке все остальные столбцы актуальны на начало операции 3
dftmp['deltaTime_34']=dftmp.apply(lambda x: x.date_op_x_-x.date_op_x,axis=1)
dftmp
```
